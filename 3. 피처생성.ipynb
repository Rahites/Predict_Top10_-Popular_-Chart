```python
# 한국어 불용어
with open('stopword.txt', 'r', encoding='utf-8') as f:
    text = f.read()

stopwords = text.split()    
```

### 빈도 분석

```python
# Top 30 노래 가사에서 등장한 단어의 빈도
# Komoran
# 가장 많이 등장한 Top10 단어들의 빈도를 Feature로 사용
from collections import Counter

mostwords_1990s = []
mostwords_2000s = []
mostwords_2010s = []
mostwords_2020s = []

for document in noun2_1990s_sw:
    mostwords_1990s.append([y for x,y in Counter(document).most_common(10)])
for document in noun2_2000s_sw:
    mostwords_2000s.append([y for x,y in Counter(document).most_common(10)])
for document in noun2_2010s_sw:
    mostwords_2010s.append([y for x,y in Counter(document).most_common(10)])
for document in noun2_2020s_sw:
    mostwords_2020s.append([y for x,y in Counter(document).most_common(10)])    
```

```python
import random
from gensim.models import word2vec
from tqdm import tqdm
```

```python
def oversample(x, n):
    lst = []
    for i in x:
        tmp = []
        for j in range(n):
            tmp += i
        lst.append(tmp)
    return lst
```

```python
w2v_noun_1990s = []
w2v_noun_2000s = []
w2v_noun_2010s = []
w2v_noun_2020s = []

for i in noun2_1990s_sw:
    w2v_noun_1990s.append("/".join(i))
for i in noun2_2000s_sw:
    w2v_noun_2000s.append("/".join(i))
for i in noun2_2010s_sw:
    w2v_noun_2010s.append("/".join(i))
for i in noun2_2020s_sw:
    w2v_noun_2020s.append("/".join(i))  
```

```python
w2v_input_90 = oversample(w2v_noun_1990s, 3)
w2v_input_00 = oversample(w2v_noun_2000s, 3)
w2v_input_10 = oversample(w2v_noun_2010s, 3)
w2v_input_20 = oversample(w2v_noun_2020s, 3)
```

```python
w2v_90 = word2vec.Word2Vec(sentences = w2v_input_90,
                        vector_size = 30,
                        window = 3,
                        min_count = 1,
                        sg = 1).wv
w2v_00 = word2vec.Word2Vec(sentences = w2v_input_00,
                        vector_size = 30,
                        window = 3,
                        min_count = 1,
                        sg = 1).wv
w2v_10 = word2vec.Word2Vec(sentences = w2v_input_10,
                        vector_size = 30,
                        window = 3,
                        min_count = 1,
                        sg = 1).wv
w2v_20 = word2vec.Word2Vec(sentences = w2v_input_20,
                        vector_size = 30,
                        window = 3,
                        min_count = 1,
                        sg = 1).wv
```

```python
train_mean_vector_90 = []

for words in tqdm(w2v_input_90):
        tmp = np.zeros(30)        
        cnt = 0
        for word in words:
            try:
                tmp += w2v_90[word]
                cnt += 1
            except:
                pass
        tmp /= cnt                    
        train_mean_vector_90.append(tmp)
        
train_mean_vector_90 = np.array(train_mean_vector_90)
```

```python
train_mean_vector_00 = []

for words in tqdm(w2v_input_00):
        tmp = np.zeros(30)            
        cnt = 0
        for word in words:
            try:
                tmp += w2v_00[word]
                cnt += 1
            except:
                pass
        tmp /= cnt                  
        train_mean_vector_00.append(tmp)
        
train_mean_vector_00 = np.array(train_mean_vector_00)
```

```python
train_mean_vector_10 = []

for words in tqdm(w2v_input_10):
        tmp = np.zeros(30)            
        cnt = 0
        for word in words:
            try:
                tmp += w2v_10[word]
                cnt += 1
            except:
                pass
        tmp /= cnt                    
        train_mean_vector_10.append(tmp)
        
train_mean_vector_10 = np.array(train_mean_vector_10)
```

```python
train_mean_vector_20 = []

for words in tqdm(w2v_input_20):
        tmp = np.zeros(30)            
        cnt = 0
        for word in words:
            try:
                tmp += w2v_20[word]
                cnt += 1
            except:
                pass
        tmp /= cnt                     
        train_mean_vector_20.append(tmp)
        
train_mean_vector_20 = np.array(train_mean_vector_20)
```
